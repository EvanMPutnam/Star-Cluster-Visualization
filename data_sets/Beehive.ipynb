## Beehive Star Cluster Visualization

https://gea.esac.esa.int/archive/

SQL:

SELECT * from gaiadr2.gaia_source where 
RA between 128 and 132 AND
DEC between 18 and 22 AND
parallax > 0 AND parallax < 15 AND
phot_g_mean_mag < 18

```python
import math
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D

def distance(x):
    #it shouldn't be less than 0 since I'm not getting any negative parallaxes from Gaia
    if x < 0:
        return 0
    else:
        return 1/x
def x(R, b, l):
    return R * math.cos(b) * math.cos(l)

def y(R, b, l):
    return R * math.cos(b) * math.sin(l)

def z(R, b):
    return R * math.sin(b)
```

```python
df = pd.read_csv("beehive.csv")

df.loc[:, 'parallax_arcsec'] = df['parallax'].apply(lambda x: x*.001)
df.loc[:, 'R'] = df['parallax_arcsec'].apply(distance)
df.loc[:, 'x'] = df.apply(lambda r: x(r['R'], r['b'], r['l']), axis=1)
df.loc[:, 'y'] = df.apply(lambda r: y(r['R'], r['b'], r['l']), axis=1)
df.loc[:, 'z'] = df.apply(lambda r: z(r['R'], r['b']), axis=1)

fig = plt.figure()
ax = fig.add_subplot(221, projection ='3d')
ax.scatter(df['x'], df['z'], df['y'], s=0.1)
ax.set_xlim([-10000, 10000])
ax.set_ylim([-10000, 10000])
ax.set_zlim([-10000, 10000])
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_zlabel('y')
ax.set_title('Beehive')

ax1 = fig.add_subplot(222, projection ='3d')
ax1.scatter(df['x'], df['z'], df['y'], s=0.1)
ax1.set_xlim([-10000, 10000])
ax1.set_ylim([-10000, 10000])
ax1.set_zlim([-10000, 10000])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_zlabel('y')
ax1.set_title('Beehive')
ax1.view_init(0, 90)

ax2 = fig.add_subplot(223, projection ='3d')
ax2.scatter(df['x'], df['z'], df['y'], s=0.1)
ax2.set_xlim([-10000, 10000])
ax2.set_ylim([-10000, 10000])
ax2.set_zlim([-10000, 10000])
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_zlabel('y')
ax2.set_title('Beehive')
ax2.view_init(0, 180)

ax3 = fig.add_subplot(224, projection ='3d')
ax3.scatter(df['x'], df['z'], df['y'], s=0.1)
ax3.set_xlim([-10000, 10000])
ax3.set_ylim([-10000, 10000])
ax3.set_zlim([-10000, 10000])
ax3.set_xlabel('x')
ax3.set_ylabel('z')
ax3.set_zlabel('y')
ax3.set_title('Beehive')
ax3.view_init(-90, 0)

plt.subplots_adjust(0.4, 0.4, 2.3, 2.3)
plt.show()
```
